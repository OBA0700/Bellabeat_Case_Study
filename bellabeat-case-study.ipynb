## Dataset
**[FitBit Fitness Tracker Data](https://www.kaggle.com/datasets/arashnic/fitbit)
 (CC0: Public Domain; made available through Mobius)** is a Kaggle dataset which contains personal fitness track from specific fitbit users. These eligible Fitbit users consented to the submission of personal tracked data, including minute-level output for physical activity, heart rate, and sleep monitoring.<br>
 It includes information about daily activity, steps, and heart rate that can be used to explore users’ habits.

## Data Limitations and Integrity
* **Data Source:** The dataset was generated by respondents to a distributed survey via Amazon Mechanical Turk between 03.12.2016-05.12.2016, which made it an open-source data. It, then so, is more like a third party data as neither was it collated *within* the Bellabeat nor Fitbiz companies’ activity paths.<br>
* **Consistency:** The data is not consistent at the point of juxtaposition of the sub-level activities for summary with the main activity levels. Speaking of the sum of steps in minute and hour compare to that of daily.
* **Period:** The data is not updated as it was recorded based on the activities back in the year 2016. There may have been some change in the condition of the resulting factors of the activity tracks within the years space.

## With "D.A.S_FormattedActivityDate" dataframe:
### i. Calculate the Overall amount of Calories burnt per activity days, Group by ActivityDate, sorted by sum_Calories in a Descending order.

```r
TotalCalories_Per_Day_Overall_IDs <- D.A.S_FormattedActivityDate %>%
  group_by(ActivityDate) %>%
  summarize(., sum_Calories = sum(Calories)) %>% arrange(desc(sum_Calories))
TotalCalories_Per_Day_Overall_IDs
```

### ii. Calculate the Mean Value of the Overall amount of Calories burnt per activity days, Grouped by ActivityDate, sorted by mean_Calories in a Descending order.

```r
MeanCalories_Overall_IDs <- D.A.S_FormattedActivityDate %>%
  group_by(ActivityDate) %>%
  summarize(., mean_Calories = mean(Calories)) %>% arrange(desc(mean_Calories))
MeanCalories_Overall_IDs
```

### iii. Calculate the Mean Value of the Overall amount of Calories burnt

```r
MeanCalories <- D.A.S_FormattedActivityDate %>%
  summarize(., mean_Calories = mean(Calories))
MeanCalories
```

Note: This presents the rate of Calorie-burn by each Participant in the overall Activity session

### iv. Calculate the Total Calories Burnt by each Participant, Group by the session ID, sorted by sum_Calories in a Descending order.

```r
TotalCalories_Per_ID <- D.A.S_FormattedActivityDate %>%
  group_by(Id) %>%
  summarize(., sum_Calories = sum(Calories), sum_TotalDistance.by.Tracker= sum(TotalDistance.by.Tracker))
TotalCalories_Per_ID
```
